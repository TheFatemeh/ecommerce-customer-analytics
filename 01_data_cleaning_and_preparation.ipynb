**Goal:** Analyze e-commerce data to identify opportunities to increase repeat purchases and revenue.

# Loading Data

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("olistbr/brazilian-ecommerce")

print("Path to dataset files:", path)
```

```python
import pandas as pd

# Load all CSV files from the dataset
orders = pd.read_csv(f'{path}/olist_orders_dataset.csv')
customers = pd.read_csv(f'{path}/olist_customers_dataset.csv')
order_items = pd.read_csv(f'{path}/olist_order_items_dataset.csv')
products = pd.read_csv(f'{path}/olist_products_dataset.csv')
order_reviews = pd.read_csv(f'{path}/olist_order_reviews_dataset.csv')

print("Dataset files loaded successfully!")
print(f"Orders shape: {orders.shape}")
print(f"Customers shape: {customers.shape}")
print(f"Order items shape: {order_items.shape}")
print(f"Products shape: {products.shape}")
print(f"Order reviews shape: {order_reviews.shape}")
```

```python
# Convert timestamp columns to datetime in all tables
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
orders['order_approved_at'] = pd.to_datetime(orders['order_approved_at'])
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'])
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'])
orders['order_estimated_delivery_date'] = pd.to_datetime(orders['order_estimated_delivery_date'])

order_reviews['review_creation_date'] = pd.to_datetime(order_reviews['review_creation_date'])
order_reviews['review_answer_timestamp'] = pd.to_datetime(order_reviews['review_answer_timestamp'])

order_items['shipping_limit_date'] = pd.to_datetime(order_items['shipping_limit_date'])

print("Timestamps converted successfully!")
```

```python
# Print all tables and their columns
tables = {
    'orders': orders.columns.tolist(),
    'customers': customers.columns.tolist(),
    'order_items': order_items.columns.tolist(),
    'products': products.columns.tolist(),
    'order_reviews': order_reviews.columns.tolist()
}

for table_name, columns in tables.items():
    print(f"\n{table_name}:")
    for col in columns:
        print(f"  - {col}")
```

# Merging Tables


## Orders + Customers Table

```python
df_orders_customers = orders.merge(customers, on="customer_id", how="left")
df_orders_customers.head()
```

## Order-level revenue Table

```python
order_revenue = order_items.groupby("order_id").agg({
    "price": "sum",
    "freight_value": "sum"
}).reset_index()

df_orders_full = orders.merge(order_revenue, on="order_id", how="left")
df_orders_full = df_orders_full.merge(order_reviews, on="order_id", how="left")
df_orders_full = df_orders_full.merge(customers[['customer_id', 'customer_unique_id']], on="customer_id", how="left")
df_orders_full.head()
```

## Product/Category Analysis

```python
df_items_products = order_items.merge(products, on="product_id", how="left")
df_items_products.head()    
```

```python
# Print columns for the 3 new merged tables
new_tables = {
    'df_orders_customers': df_orders_customers.columns.tolist(),
    'df_orders_full': df_orders_full.columns.tolist(),
    'df_items_products': df_items_products.columns.tolist()
}

for table_name, columns in new_tables.items():
    print(f"\n{table_name}:")
    for col in columns:
        print(f"  - {col}")
```

```python
import os

os.makedirs('data/processed', exist_ok=True)

df_items_products.to_csv('data/processed/df_items_products.csv', index=False)
df_orders_full.to_csv('data/processed/df_orders_full.csv', index=False)
df_orders_customers.to_csv('data/processed/df_orders_customers.csv', index=False)

print("All dataframes saved successfully!")
```
